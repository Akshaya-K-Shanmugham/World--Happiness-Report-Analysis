```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python

df = pd.read_csv("C:/Users/LENOVO/Documents/progrms/World  happiness report/world-happiness-report-2021.csv")
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
print("\nColumn Names:")
print(df.columns)
```

```python
if df.duplicated().sum() > 0:
    print("\nRemoving duplicates...")
    df.drop_duplicates(inplace=True)


```

```python

print(df.columns.tolist())
```

```python

df.rename(columns={
    "Country name": "Country",
    "Ladder score": "Happiness Score", 
    "Logged GDP per capita": "GDP", 
    "Social Support": "Social Support", 
    "Life Expectancy": "Life Expectancy", 
    "Freedom": "Freedom", 
    "Generosity": "Generosity", 
    "Corruption": "Corruption"
}, inplace=True)


print(df.columns.tolist())
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df['Happiness Score'], kde=True, color='blue')
plt.title("Distribution of Happiness Scores", fontsize=16)
plt.xlabel("Happiness Score", fontsize=12)
plt.show()
```

```python

numeric_df = df.select_dtypes(include=['float64', 'int64'])


corr_matrix = numeric_df.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python


print(df.columns.tolist())
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="GDP", y="Happiness Score", data=df, hue="Regional indicator", palette="viridis")
plt.title("Happiness Score vs GDP per Capita", fontsize=16)
plt.xlabel("GDP per Capita", fontsize=12)
plt.ylabel("Happiness Score", fontsize=12)
plt.show()
```

```python
top_10 = df.nlargest(10, 'Happiness Score')
plt.figure(figsize=(10, 6))
sns.barplot(y="Country", x="Happiness Score", data=top_10, palette="viridis")
plt.title("Top 10 Happiest Countries", fontsize=16)
plt.xlabel("Happiness Score", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.show()
```

```python
print("\nInsights:")
print(f"1. Average Happiness Score: {df['Happiness Score'].mean():.2f}")
print(f"2. Countries with the highest GDP tend to have higher Happiness Scores.")
print(f"3. The strongest positive correlation is between GDP and Happiness Score ({corr_matrix['Happiness Score']['GDP']:.2f}).")
print(f"4. Countries with high Social Support also show higher Happiness Scores.")
```
